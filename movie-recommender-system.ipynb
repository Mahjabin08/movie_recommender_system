```python
#it threw a error as it's a string
#so we will use a library to convert it to list of strings
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

## Vectorization

Now, we will merge all the tags, and generate a large text, and choose the 5000 most common words among them,and then match the tags of the movies and count the occurance of those words in those tags, now we have a data frame of shape (5000,5000).
To simplify things, let's assume, we choose only two words instead of 5000,for example action and future, so this will be a two dimensional space, and the five thousand movies will have their place in the 2d space as shown below:

now think of it with 5000 words, in 5000 dimensional space, this process is vecorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors[0]
```

```python
len(cv.get_feature_names_out())
```

```python
import nltk

```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    #returning y as a string    
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

According to Curse of dimensionality, Euclidean distance is not that of a reliable measure, on the other hand, cosine distance is a much better choice. Because in high dimensional vectors, there could be much noise, and redundancy, and measuring the magnitude of the vectors would produce irrelavant distance to the work, and deviate the results from the focus. But in case of cosine similarity, it just calculates the angle between two vectors, so it puts focus on the direction rather than the magnitud, in turn, producing a much precise and relavent result. As we are working with high dimensional vectors here, we are going to use cosine distance to find out the similarity here between the movies. The more the distance between movies, the more different they are.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie_similarity=cosine_similarity(vectors)
```

```python
#it is showing the similarity of the first movie with the all the 4806 movies
movie_similarity[0]
```

```python
#making a list of tuples for keeping track of the indexing, so that it stores the order when we sort it
list_tuples=list(enumerate(movie_similarity[0]))
#now sorting based on the second element of the list_tuples
sorted(list_tuples, reverse=True,key=lambda x:x[1])[1:6]
```

```python
#creating movie recommender function

def movie_recommend(movie):
    movie_index = new_df[new_df['title']==movie].index[0]
    movie_distances = movie_similarity[movie_index]
    movie_list= sorted(list(enumerate(movie_distances)), reverse=True,key=lambda x:x[1])[1:6]
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```

```python
movie_recommend('Batman Begins')
```
